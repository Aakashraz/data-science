```python
df5 = pd.DataFrame({
    'id' : [1,2,3,4],
    'name' : ['Kreeti', 'Amin', 'Maya','Keshav']
})

df6 = pd.DataFrame({
    'id' : [3,4, 5,6],
    'course' : ['GenAI', 'LLM', 'CPP', 'HLD']
})
# inner_result = pd.merge(df5, df6, how='inner', on='id')
# outer_result = pd.merge(df5, df6, how='outer', on='id')
left_result = pd.merge(df5, df6, how='left', on='id')
right_result = pd.merge(df5, df6, how='right', on='id')

left_result, right_result
```

```python
df11 = pd.DataFrame({
    'id1' : [1,2,3,4],
    'name' : ['Kreeti', 'Amin', 'Maya','Keshav']
})

df12 = pd.DataFrame({
    'id2' : [3,4, 5,6],
    'course' : ['GenAI', 'LLM', 'CPP', 'HLD']
})
outer= pd.merge(df11, df12, how='outer', left_on='id1', right_on='id2')
outer
```

```python
frames = [df11, df12]
result = pd.concat(frames, axis=0, keys=['df11', 'df12'])
# this is example of Multilevel Indexing
# result
result.isna(), result, result.notna()
```
